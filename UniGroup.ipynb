```python
def label_states (row):
    if row['State'] in ['ME', 'NH', 'VT', 'MA', 'RI', 'CT', 'NY', 'PA', 'NJ']:
        return 'north-east'
    if row['State'] in ['WI', 'MI', 'IL', 'IN', 'OH', 'ND', 'SD', 'NE', 'KS', 'MN', 'IA', 'MO']:
        return 'midwest'
    if row['State'] in ['DE', 'MM', 'DC', 'VA', 'WV', 'NC', 'SC', 'GA', 'FL', 'KY', 'TN', 'MS', 'AL', 'OK', 'TX', 'AR', 'LA']:
        return 'south'
    if row['State'] in ['MT', 'ID', 'WY', 'CO', 'NM', 'AZ', 'UT', 'NV', 'CA', 'OR', 'WA', 'AK',  'HI']:
        return 'west'
    
unidf['region'] = unidf.apply(lambda row: label_states(row), axis=1)
unidf    
```

```python
statscounts = unidf.groupby(["Prof G Categorization", 'region'])['Institution Name']
statscounts.count()
sc = statscounts.count()
sc = sc.to_frame()
sc
```

```python
#data = sc.set_index(["Prof G Categorization", 'State']).value

sc.unstack().plot(kind='bar', stacked=True)
```

```python
fig, ax = plt.subplots(figsize=(10,7))  

sta = sc['State']
margin_bottom = np.zeros(len(sc['Prof G Categorization']))
colors = ["#006D2C", "#31A354","#74C476"]

for num, Stat in enumerate(sta):
    values = list(sc[sc['State'] == Stat].loc[:, 'Institution Name'])

    sc[sc['State'] == Stat].plot.bar(x='Prof G Categorization',y='Institution Name', ax=ax, stacked=True, 
                                    bottom = margin_bottom, color=colors[num], label=Stat)
    margin_bottom += values

plt.show()
```
